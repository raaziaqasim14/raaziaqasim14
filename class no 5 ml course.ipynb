```python
#loading libraries 
import matplotlib.pyplot as plt
import seaborn as sns
```

### Line plot in Matplotlib:

```python
# taking data
x = [2,4,6,8,10]
y = [-2,-4,-6,-8,-10]
```

```python
# plotting the graph
plt.plot(x,y)
```

```python
# refining the line plot
plt.plot(x,y)
plt.title('graph of a linear eq')
plt.xlabel('x-values')
plt.ylabel('y-values')

plt.show
```

```python
# making the graph more eye pleasing 

plt.plot(x,y,color='orange',linestyle = '-.', label = 'sol of eq')
plt.legend()
plt.title('graph of a linear equation')
plt.xlabel('x-values')
plt.ylabel('y-values')

plt.show()
```

### Scatter plot in Matplotlib:

```python
#scatter plot
plt.scatter(x,y,color='brown',marker='*')
plt.title("scatter plot")
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

### Multiple plots using matplotlib:

```python
plt.figure(figsize=(14,6)) #canvas size 

plt.subplot(2,1,1) # 2 rows, 1 column, 1st graph 
plt.plot(x,y)
plt.title("1st plot")

plt.subplot(2,1,2) # 2 rows, 1 column, 2nd graph 
plt.scatter(x,y)
plt.title("2nd plot")

plt.tight_layout() #overlapping kthm hojati hay
plt.show
```

```python
plt.figure(figsize=(14,6)) #canvas size 

plt.subplot(1,2,1) # 2 rows, 1 column, 1st graph 
plt.plot(x,y)
plt.title("1st plot")

plt.subplot(1,2,2) # 2 rows, 1 column, 2nd graph 
plt.scatter(x,y)
plt.title("2nd plot")

plt.tight_layout() #overlapping kthm hojati hay
plt.show()
```

### Bar plot in Matplotlib

```python
# sample data
khanay = ['alu gobi','pasta','pizza','bhindi']
tadad = [2, 14, 1, 5]

#creating a bar plot
plt.bar(khanay,tadad)

#adding title and labels
plt.xlabel("KHANAY")
plt.ylabel("No. of plates")
plt.title("RAAT KA KHANA")

plt.show()
```

### Seaborn Barplot:

```python
# tips dataset
tips = sns.load_dataset('tips')

tips.head()
```

```python
sns.barplot(y='tip',x='day',data=tips)
plt.title("Colorful Din")
plt.show()
```

```python
# seaborn scatter plot
sns.scatterplot(x='tip',y='total_bill',data=tips)
```

```python
# count plot
## kitnay adami thay...
print(tips['sex'].value_counts())
sns.countplot(x='sex',data=tips)

sns.despine() # upr or neechay line khtm 
plt.show()
```

```python
# count plot
## kitnay adami thay...
print(tips['sex'].value_counts())
sns.countplot(x='sex',data=tips,palette='dark', hue='smoker')

sns.despine() # upr or neechay line khtm 
plt.show()
```

```python
sns.catplot(x='sex',hue = 'smoker', col = 'day', data = tips, kind = 'count')
```

```python
#sns.set_theme(style="whitegrid")


# Draw a nested violinplot and split the violins for easier comparison
sns.violinplot(data=tips, x="day", y="total_bill", hue="smoker",
               split=True, inner="quart", linewidth=1,
              palette={"Yes": "b", "No": ".85"})
sns.despine(left=True)
```
